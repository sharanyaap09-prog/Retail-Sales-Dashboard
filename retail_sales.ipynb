```python
#Analysing Data
```

```python
sales['YearMonth'] = sales['OrderDate'].dt.to_period('M')
```

```python
monthly_sales = sales.groupby('YearMonth')['Sales'].sum().reset_index()

monthly_sales
```

```python
monthly_sales['YearMonth'] = monthly_sales['YearMonth'].astype(str)

plt.figure(figsize=(12,5))
plt.plot(monthly_sales['YearMonth'], monthly_sales['Sales'])
plt.xticks(rotation=45)
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
region_analysis = sales.groupby('Region')[['Sales','Profit']].sum()

region_analysis.sort_values('Sales', ascending=False)
```

```python
region_analysis.plot(
    kind='bar',
    figsize=(10,5)
)

plt.title('Sales and Profit by Region')
plt.ylabel('Amount')
plt.show()
```

```python
category_sales = sales.groupby('Category')['Sales'].sum()

category_sales.sort_values(ascending=False)
```

```python
category_sales.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Sales by Category')
plt.ylabel('Sales')
plt.show()
```

```python
subcategory_sales = sales.groupby('SubCategory')['Sales'].sum()

subcategory_sales = subcategory_sales.sort_values(
    ascending=False
)

subcategory_sales
```

```python
subcategory_sales.head(10).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title('Top 10 Sub-Categories by Sales')
plt.ylabel('Sales')
plt.show()
```

```python
sales['Year'] = sales['OrderDate'].dt.year
yearly_sales = sales.groupby('Year')['Sales'].sum()

yearly_sales
```

```python
yoy_growth = yearly_sales.pct_change() * 100

yoy_growth = yoy_growth.round(2)

print(yoy_growth)
```

```python
yoy_table = pd.DataFrame({
    'Sales': yearly_sales,
    'YoY Growth %': yoy_growth
})

yoy_table
```

```python
yearly_sales.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Yearly Sales')
plt.ylabel('Sales')
plt.show()
```

```python
# Extract year
sales['Year'] = sales['OrderDate'].dt.year

# Calculate yearly sales
yearly_sales = sales.groupby('Year')['Sales'].sum().reset_index()

# Calculate YoY Growth %
yearly_sales['YoY_Growth_%'] = (
    yearly_sales['Sales'].pct_change() * 100
).fillna(0).round(2)

# Create line chart for YoY Growth
fig = px.line(
    yearly_sales,
    x='Year',
    y='YoY_Growth_%',
    markers=True,
    title='Year-over-Year Sales Growth (%)'
)

fig.show()
```

```python
#KPI Calculations
```

```python
total_revenue = sales['Sales'].sum()

total_profit = sales['Profit'].sum()

profit_margin = (total_profit / total_revenue) * 100

print("Total Revenue:", round(total_revenue,2))
print("Total Profit:", round(total_profit,2))
print("Profit Margin (%):", round(profit_margin,2))
```

```python
total_revenue = sales['Sales'].sum()

total_orders = sales['OrderID'].nunique()

aov = total_revenue / total_orders

print("Average Order Value:", round(aov,2))
```

```python
top_customers = (
    sales.groupby('CustomerID')['Sales']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(top_customers)
```

```python
returned_orders = sales[sales['Returned'] == 'Yes'].shape[0]

total_orders = len(sales)

return_rate = (returned_orders / total_orders) * 100

print("Return Rate (%):", round(return_rate,2))
```

```python
customer_orders = sales.groupby('CustomerID')['OrderID'].nunique()

repeat_customers = customer_orders[customer_orders > 1].count()

total_customers = customer_orders.count()

repeat_customer_rate = (repeat_customers / total_customers) * 100

print("Repeat Customer Rate (%):", round(repeat_customer_rate,2))
```

```python
#Final KPI Summary Table
summary = pd.DataFrame({
    'Metric': [
        'Total Revenue',
        'Total Profit',
        'Profit Margin (%)',
        'Average Order Value',
        'Return Rate (%)',
        'Repeat Customer Rate (%)'
    ],
    'Value': [
        round(total_revenue,2),
        round(total_profit,2),
        round(profit_margin,2),
        round(aov,2),
        round(return_rate,2),
        round(repeat_customer_rate,2).T
    ]
})

summary
```
